Drop missing values in columns that will be used in the project such as: **EdLevel, Industry, RemoteWork, YearsCodePro, LanguageHaveWorkedWith, LanguageWantToWorkWith, LanguageAdmired, DatabaseAdmired, WebframeHaveWorkedWith, AIToolCurrently Using**.

Instead of dropping, it will be used fill in the missing values in the columns like **CompTotal, ConvertedCompYearly, JobSat**.

```python
df.dropna(subset = [
    "EdLevel", 
    "Industry", 
    "RemoteWork", 
    "YearsCodePro",
    "LanguageHaveWorkedWith",
    "LanguageWantToWorkWith",
    "LanguageAdmired",
    "DatabaseAdmired",
    "WebframeHaveWorkedWith",
    "AIToolCurrently Using"], inplace=True)

print(df.shape)
```

```python
median_ct = df["CompTotal"].median()
mean_ct = df["CompTotal"].mean()

print(f"Median of CompTotal: {median_ct}")
print(f"Mean of CompTotal: {mean_ct}")
```

```python
df["CompTotal"] = df["CompTotal"].fillna(mean_ct)
```

```python
median_ccy = df["ConvertedCompYearly"].median()
mean_ccy = df["ConvertedCompYearly"].mean()

print(f"Median of ConvertedCompYearly: {median_ccy}")
print(f"Mean of ConvertedCompYearly: {mean_ccy}")
```

```python
df["ConvertedCompYearly"] = df["ConvertedCompYearly"].fillna(mean_ccy)
```

```python
median_js = df["JobSat"].median()
mode_js = df["JobSat"].mode()[0]
mean_js = df["JobSat"].mean()

print(f"Median of JobSat: {median_js}")
print(f"Mode of JobSat: {mode_js}")
print(f"Mean of JobSat: {mean_js}")
```

```python
df["JobSat"] = df["JobSat"].fillna(mode_js)
```

Also, **Country** column will be used in **IBM Cognos Analytics**. So, the column needs to be standardized.

```python
df["Country"] = df["Country"].replace({
    "Republic of Korea": "South Korea",
    "Democratic People's Republic of Korea": "North Korea",
    "Iran, Islamic Republic of...": "Iran",
    "Venezuela, Bolivarian Republic of...": "Venezuela",
    "Libyan Arab Jamahiriya": "Libya",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "Micronesia, Federated States of...": "Micronesia",
    "Congo, Republic of the...": "Republic of the Congo"
})

df = df[~df["Country"].isin(["Nomadic"])]
df = df.dropna(subset=["Country"])
```

## Q1: What is the age distribution?

```python
df["Age"].unique()
```

```python
def convert_ages_to_numeric(age):
    if age == "Under 18 years old":
        return 17
    elif age == "18-24 years old":
        return 21
    elif age == "25-34 years old":
        return 29
    elif age == "35-44 years old":
        return 39
    elif age == "45-54 years old":
        return 49
    elif age == "55-64 years old":
        return 59
    elif age == "65 years or older":
        return 65
    else:
        return np.nan
    
df["Age_numeric"] = df["Age"].apply(convert_ages_to_numeric)

print(df[["Age", "Age_numeric"]].head())
```

```python
print("Median age: {}".format(df["Age_numeric"].median()))
print("Mode age {}".format(df["Age_numeric"].mode()[0]))
print("Mean age: {}".format(df["Age_numeric"].mean()))
```

```python
# plt.figure(figsize=(8,4))
sns.countplot(data=df, x="Age", hue="Age", palette="Blues_r")
plt.title("Age Distribution", fontsize=20)
plt.xlabel("Age ranges")
plt.ylabel("No.")
plt.xticks(rotation=30)
plt.tight_layout()
```

## Q2: What are the most common education levels among employees?

```python
df["EdLevel"].unique()
```

```python
def simplify_ed(ed):
    if ed == "Bachelor’s degree (B.A., B.S., B.Eng., etc.)":
        return "Bachelor's degree"
    elif ed == "Primary/elementary school":
        return "Primary/elementary school"
    elif ed == "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)":
        return "Master's degree"
    elif ed == "Some college/university study without earning a degree":
        return "Some college, no degree"
    elif ed == "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)":
        return "Secondary school"
    elif ed == "Professional degree (JD, MD, Ph.D, Ed.D, etc.)":
        return "Professional degree"
    elif ed == "Associate degree (A.A., A.S., etc.)":
        return "Associate degree"
    else:
        return "Other"

df["EdLevel_simplified"] = df["EdLevel"].apply(simplify_ed)

print(df[["EdLevel", "EdLevel_simplified"]].head())
```

```python
ed_level_counts = df["EdLevel_simplified"].value_counts()
ed_level_counts_df = ed_level_counts.reset_index()
ed_level_counts_df.columns = ["Education Level", "No. of Respondents"]

plt.figure(figsize=(14, 5))
sns.barplot(data=ed_level_counts_df, x="Education Level", y="No. of Respondents", hue="Education Level", palette="Reds_r")
plt.title("Most Common Education Level of Respondents", fontsize=20)
plt.tight_layout()
```

## Q11: How do salaries vary by work experience?

```python
df_corr = df.corr(numeric_only=True).apply(lambda col: col.map(lambda x: f"{x:.4f}"))
df_corr["ConvertedCompYearly"].sort_values(ascending=False)
```

```python
corr_coeff = df.corr(numeric_only=True).loc["WorkExp", "ConvertedCompYearly"]

g = sns.relplot(data=df, x="WorkExp", y="ConvertedCompYearly", kind="line")

ax = g.ax  

ax.set_title("Yearly Compensation by Work Experience", fontsize=20)
ax.set_ylabel("Converted Compensation Yearly (million)")
ax.set_xlabel("Work Experience (year)")

ax.legend([f"Correlation: {corr_coeff:.2f}"], loc="upper left", fontsize=12)

plt.tight_layout()
```
